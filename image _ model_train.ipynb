Import Data

```python
import numpy as np
import cv2
import os
import glob
import matplotlib.pyplot as plt
import pandas as pd
```

```python
img_dir = r"/content/male"
data_path = os.path.join(img_dir,'*g')
files = glob.glob(data_path)
data = []
for m in files:
  img = cv2.imread(m)
  img.resize((200,200))
  data.append(img)
```

```python
img_dir = r"/content/female"
data_paths = os.path.join(img_dir,'*g')
filess = glob.glob(data_paths)
data2 = []
for f in filess:
  img2 = cv2.imread(f)
  img2.resize((200,200))
  data2.append(img2)
```

```python
img_dir = r"/content/kids"
data_pathss = os.path.join(img_dir,'*g')
filesss = glob.glob(data_pathss)
data3 = []
for k in filesss:
  img3 = cv2.imread(k)
  img3.resize((200,200))
  data3.append(img3)
```

```python
data2
```

```python
data3
```

 Preprocessing

```python
x = data+data2+data3
x
```

```python
x= np.array(x)
```

```python
x.shape
```

```python
x = x.reshape(60,200*200)
```

```python
x.shape
```

```python
o = np.zeros([20])
o
```

```python
z = np.ones([20])
z
```

```python
z1 = []
for i in range(1,21):
  z1.append(2)
```

```python
z1
```

```python
z1=np.array(z1)
z1
```

```python
y = np.append(o,z)
```

```python
y = np.append(y,z1)
```

```python
y
```

```python
y.shape
```

Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest , ytrain, ytest = train_test_split(x,y , test_size = 0.1 ,shuffle = True)
```

SVM

```python
from sklearn import svm
```

```python
S = svm.SVC()
S = S.fit(xtrain, ytrain)
```

```python
pred = S.predict(xtest)
```

```python
accuracy = S.score(xtest , ytest)
```

```python
accuracy
```

SVM Evaluation

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(ytest,pred)
```

```python
cm
```
